# SynapseML LightGBM — Clean & Cluster-Optimized Notebook

Notebook này dùng để phát hiện **data drift/anomaly theo batch** bằng cách:

- `label = 1`: batch hiện tại cần kiểm tra.
- `label = 0`: các batch lịch sử dùng để so sánh.
- Train **SynapseML LightGBM** để phân biệt 2 nhóm dữ liệu.
- Dùng **AUC** để đo mức độ khác nhau giữa batch hiện tại và lịch sử.

Thiết kế vẫn giữ các tối ưu chính của notebook cũ:

1. Không dùng `orderBy(rand()).limit()` trên toàn bộ batch lớn.
2. Sampling theo `processing_date_hour` bằng candidate sampling trước, rồi mới random sort + limit.
3. Spark split theo label bằng hash/window để gần giống `train_test_split(..., stratify=y)`.
4. Cache/persist các DataFrame dùng lại nhiều lần.
5. SynapseML profile tối ưu cho cụm `2 executors x 6 cores`, `task.cpus = 3`, `numTasks = 4`, `numThreads = 3`.
6. Không bật SHAP trong Optuna vì rất nặng; chỉ bật ở final model nếu cần.

## 4. Load Silver data

Ở đây ta normalize tên cột về lowercase để khớp với job Silver mới của bạn. Các cột thời gian được cast về timestamp để filter batch ổn định hơn.


```python
def normalize_spark_columns_to_lower(input_df):
    """
    Chuẩn hóa tên cột về lowercase.

    Lý do:
    - File raw/silver cũ có thể dùng VendorID, RatecodeID, PULocationID...
    - Job silver mới của bạn dùng vendorid, ratecodeid, pulocationid...
    - Normalize một lần giúp các cell phía sau không bị mismatch tên cột.
    """
    new_names = [str(c).strip().lower() for c in input_df.columns]

    if len(new_names) != len(set(new_names)):
        raise ValueError(
            "Sau khi lowercase, có cột bị trùng tên. Hãy kiểm tra schema input. "
            f"Columns after lower: {new_names}"
        )

    return input_df.select([
        F.col(old).alias(new)
        for old, new in zip(input_df.columns, new_names)
    ])


TIMESTAMP_COLUMNS = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime",
    "date_hour",
    "processing_date_hour",
]


def cast_timestamp_columns(input_df):
    """Cast các cột thời gian quan trọng về timestamp nếu chúng tồn tại."""
    out_df = input_df
    for c in TIMESTAMP_COLUMNS:
        if c in out_df.columns:
            out_df = out_df.withColumn(c, F.to_timestamp(F.col(c)))
    return out_df


df_raw = spark.read.parquet(DATA_PATH)
df_raw = normalize_spark_columns_to_lower(df_raw)
df_raw = cast_timestamp_columns(df_raw)

required_cols = {"processing_date_hour", "date_hour"}
missing_required = sorted(required_cols - set(df_raw.columns))
if missing_required:
    raise ValueError(f"Missing required columns in Silver parquet: {missing_required}")

print("Raw row count:", df_raw.count())
df_raw.printSchema()
df_raw.show(5, truncate=False)
```

## 6. Define target and comparison batches

- Target batch: `label = 1`.
- Comparison/history batches: `label = 0`.


```python
def ts_str(ts) -> str:
    """Format timestamp theo chuẩn yyyy-MM-dd HH:mm:ss."""
    return pd.Timestamp(ts).strftime("%Y-%m-%d %H:%M:%S")


def ts_lit(ts):
    """Tạo Spark timestamp literal để so sánh với cột timestamp."""
    return F.to_timestamp(F.lit(ts_str(ts)))


def filter_processing_hour(source_df, processing_hour):
    """Lọc DataFrame theo một processing_date_hour."""
    return source_df.filter(F.col("processing_date_hour") == ts_lit(processing_hour))


def get_comparison_processing_hours(target_processing_hour):
    """
    Tạo danh sách batch lịch sử để so sánh với target.

    Gồm:
    - batch liền trước: target - 12h
    - batch lịch sử: target - 1 ngày, target - 7 ngày, target - 14 ngày, ...
    """
    target_processing_hour = pd.Timestamp(target_processing_hour)

    comparison_hours = [target_processing_hour - PREVIOUS_BATCH_OFFSET]
    comparison_hours += [target_processing_hour - offset for offset in HISTORY_OFFSETS]

    # set để bỏ trùng, sorted để log dễ đọc.
    return sorted(set(comparison_hours))


target_processing_hour = pd.Timestamp(TARGET_PROCESSING_DATE_HOUR)
comparison_processing_hours = get_comparison_processing_hours(target_processing_hour)

target_count = filter_processing_hour(df_raw, target_processing_hour).count()

print("Target processing hour, label = 1:")
print("-", target_processing_hour, "| rows:", target_count)

print("\nComparison processing hours, label = 0:")
for h in comparison_processing_hours:
    cnt = filter_processing_hour(df_raw, h).count()
    print("-", h, "| rows:", cnt)
```

## 7. Optimized sampling by processing hour

Đây là tối ưu quan trọng nhất so với cách `orderBy(rand()).limit()` trực tiếp trên toàn bộ batch.


```python
def sample_by_processing_hour_spark(
    source_df,
    processing_hour,
    label,
    max_rows,
    random_state,
    oversample_factor=1.6,
):
    """
    Lấy tối đa max_rows dòng cho một processing_date_hour và gán nhãn label.

    Tối ưu:
    - Nếu batch nhỏ hơn max_rows: lấy toàn bộ, không sample.
    - Nếu batch lớn hơn max_rows:
        1. Sample trước một candidate set nhỏ hơn.
        2. Chỉ random sort candidate rồi limit.
        3. Nếu candidate vô tình thiếu dòng, fallback về random sort toàn bộ batch.

    Vì sao nhanh hơn?
    - orderBy(rand()) là thao tác shuffle/sort rất nặng.
    - Sort 16,000 dòng candidate nhẹ hơn sort 1,000,000 dòng của batch gốc.
    """
    part = filter_processing_hour(source_df, processing_hour)
    total_rows = part.count()

    # Giữ schema + thêm label dù batch rỗng, để các bước union phía sau không lỗi.
    part = part.withColumn("label", F.lit(float(label)))

    if total_rows == 0:
        return part

    # Batch nhỏ: lấy toàn bộ để không mất dữ liệu.
    if total_rows <= max_rows:
        return part

    # Batch lớn: sample dư candidate trước.
    frac = min(
        1.0,
        float(max_rows) / float(total_rows) * float(oversample_factor)
    )

    candidate = part.sample(
        withReplacement=False,
        fraction=frac,
        seed=int(random_state),
    )

    candidate_rows = candidate.count()

    # Nếu candidate đủ lớn, chỉ sort random candidate.
    if candidate_rows >= max_rows:
        return candidate.orderBy(F.rand(int(random_state) + 17)).limit(int(max_rows))

    # Fallback: đảm bảo đủ max_rows nếu batch gốc có đủ dòng.
    return part.orderBy(F.rand(int(random_state) + 19)).limit(int(max_rows))
```

## 8. Build training dataset: target vs history


```python
# ============================================================
# Positive sample: target batch, label = 1
# ============================================================
positive_df = sample_by_processing_hour_spark(
    source_df=df_raw,
    processing_hour=target_processing_hour,
    label=1,
    max_rows=TARGET_SAMPLE_PER_GROUP,
    random_state=RANDOM_STATE,
    oversample_factor=OVERSAMPLE_FACTOR,
)

positive_count = positive_df.count()
print("Positive sample:")
print("Processing hour:", target_processing_hour)
print("Rows:", positive_count)

if positive_count < MIN_ROWS_PER_GROUP:
    raise ValueError(
        f"Positive sample quá ít: {positive_count} rows. "
        f"Cần tối thiểu {MIN_ROWS_PER_GROUP} rows."
    )

positive_df.select("date_hour", "processing_date_hour", "label").show(5, truncate=False)

# ============================================================
# Negative samples: history/comparison batches, label = 0
# ============================================================
negative_parts = []

for i, compare_hour in enumerate(comparison_processing_hours):
    part = sample_by_processing_hour_spark(
        source_df=df_raw,
        processing_hour=compare_hour,
        label=0,
        max_rows=TARGET_SAMPLE_PER_GROUP,
        random_state=RANDOM_STATE + i + 100,
        oversample_factor=OVERSAMPLE_FACTOR,
    )

    cnt = part.count()
    print("Comparison hour:", compare_hour, "| rows:", cnt)

    if cnt >= MIN_ROWS_PER_GROUP:
        negative_parts.append(part)
    else:
        print("  -> Skipped because rows < MIN_ROWS_PER_GROUP")

if len(negative_parts) == 0:
    raise ValueError("Không có đủ dữ liệu lịch sử để tạo label = 0.")

negative_df = negative_parts[0]
for p in negative_parts[1:]:
    negative_df = negative_df.unionByName(p, allowMissingColumns=True)

negative_count = negative_df.count()
print("\nNegative sample total rows:", negative_count)

negative_df.groupBy("processing_date_hour") \
    .count() \
    .withColumnRenamed("count", "row_count") \
    .orderBy("processing_date_hour") \
    .show(truncate=False)
```

```python
# Gộp positive + negative thành dataset train/evaluate.
work_df = positive_df.unionByName(negative_df, allowMissingColumns=True)

# Không random sort toàn bộ dataset ở đây.
# Split phía sau đã random theo label bằng hash/window.
work_df = (
    work_df
    .repartition(SPARK_SQL_SHUFFLE_PARTITIONS)
    .persist(StorageLevel.MEMORY_AND_DISK)
)

row_count = work_df.count()
col_count = len(work_df.columns)

print("Work dataset shape:", (row_count, col_count))

print("\nLabel count:")
work_df.groupBy("label").count().orderBy("label").show(truncate=False)

print("\nRows by label and processing_date_hour:")
work_df.groupBy("label", "processing_date_hour") \
    .count() \
    .withColumnRenamed("count", "row_count") \
    .orderBy("label", "processing_date_hour") \
    .show(truncate=False)

print("\nSample rows:")
work_df.limit(5).show(truncate=False)
```

## 10. Stratified split utilities

Spark không có `train_test_split(..., stratify=y)` như sklearn. Vì vậy cell này tạo split gần tương đương:

- Tổng số dòng test ≈ `round(TEST_SIZE * total)`.
- Tổng số dòng validation ≈ `round(VALID_SIZE * total)`.
- Tỷ lệ label trong train/valid/test gần giống nhau.
- Dùng hash theo row id để random ổn định hơn `orderBy(rand())`.


```python
def get_count_and_positive_rate(input_df, label_col="label"):
    """Trả về số dòng và tỷ lệ label=1."""
    row = input_df.select(
        F.count("*").alias("n"),
        F.avg(F.col(label_col).cast("double")).alias("positive_rate"),
    ).first()

    return int(row["n"]), float(row["positive_rate"] or 0.0)


def _approximate_stratified_allocation(counts, n_take, seed=42):
    """
    Phân bổ số dòng cần lấy cho từng label.

    Ví dụ:
    - Dataset có 40% label=1, 60% label=0.
    - Cần lấy 1,000 dòng test.
    - Hàm sẽ phân bổ xấp xỉ 400 dòng label=1 và 600 dòng label=0.
    """
    counts = [int(x) for x in counts]
    total = sum(counts)
    n_classes = sum(1 for c in counts if c > 0)

    if total <= 0:
        raise ValueError("Dataset rỗng, không thể stratified split.")

    if n_take < 0 or n_take > total:
        raise ValueError(f"n_take không hợp lệ: n_take={n_take}, total={total}")

    if n_take == 0:
        return [0 for _ in counts]

    if n_take < n_classes:
        raise ValueError(
            f"n_take={n_take} nhỏ hơn số label có dữ liệu={n_classes}; "
            "không thể đảm bảo mỗi label xuất hiện trong split."
        )

    expected = [c * n_take / total for c in counts]
    allocation = [int(math.floor(x)) for x in expected]

    # Đảm bảo mỗi label có ít nhất 1 dòng nếu label đó có dữ liệu.
    for i, c in enumerate(counts):
        if c > 0 and allocation[i] == 0:
            allocation[i] = 1

    # Nếu ép min 1 làm vượt n_take, trừ bớt từ label đang dư nhiều nhất.
    while sum(allocation) > n_take:
        candidates = [i for i, _ in enumerate(counts) if allocation[i] > 1]
        if not candidates:
            break
        idx = max(candidates, key=lambda i: allocation[i] - expected[i])
        allocation[idx] -= 1

    # Nếu còn thiếu, cộng thêm theo phần thập phân lớn nhất.
    remaining = n_take - sum(allocation)

    if remaining > 0:
        rng = random.Random(seed)
        fractions = [expected[i] - math.floor(expected[i]) for i in range(len(counts))]
        candidate_indices = list(range(len(counts)))
        rng.shuffle(candidate_indices)
        candidate_indices = sorted(candidate_indices, key=lambda i: fractions[i], reverse=True)

        while remaining > 0:
            changed = False
            for i in candidate_indices:
                if remaining <= 0:
                    break
                if allocation[i] < counts[i]:
                    allocation[i] += 1
                    remaining -= 1
                    changed = True
            if not changed:
                break

    if sum(allocation) != n_take:
        raise RuntimeError(
            f"Allocation lỗi: sum(allocation)={sum(allocation)} != n_take={n_take}. "
            f"allocation={allocation}, counts={counts}"
        )

    return allocation


def stratified_split_sklearn_like_spark(
    input_df,
    label_col="label",
    test_size=0.15,
    valid_size=0.15,
    seed=42,
    cache_result=True,
    return_split_df=False,
):
    """
    Tạo train/valid/test split có stratify theo label.

    Lưu ý:
    - Hàm này có dùng window theo label, nên vẫn có shuffle.
    - Tuy nhiên nó tránh random sort toàn bộ dataset không cần thiết.
    """
    spark_session = input_df.sparkSession

    n_total = input_df.count()
    n_test = int(round(test_size * n_total))
    n_valid = int(round(valid_size * n_total))

    if n_total <= 0:
        raise ValueError("DataFrame rỗng, không thể split.")
    if n_test <= 0:
        raise ValueError(f"n_test quá nhỏ: {n_test}")
    if n_valid <= 0:
        raise ValueError(f"n_valid quá nhỏ: {n_valid}")
    if n_test + n_valid >= n_total:
        raise ValueError(
            f"Split không hợp lệ: n_test + n_valid = {n_test + n_valid}, n_total = {n_total}"
        )

    base_df = (
        input_df
        .withColumn(label_col, F.col(label_col).cast("double"))
        .withColumn("_split_row_id", F.monotonically_increasing_id())
    )

    count_rows = base_df.groupBy(label_col).count().orderBy(label_col).collect()
    labels = [float(r[label_col]) for r in count_rows]
    label_counts = [int(r["count"]) for r in count_rows]

    if len(labels) < 2:
        raise ValueError(f"Cần ít nhất 2 label để tính AUC. Hiện tại chỉ có: {labels}")

    test_alloc = _approximate_stratified_allocation(label_counts, n_test, seed=seed)

    remaining_counts = [
        label_counts[i] - test_alloc[i]
        for i in range(len(label_counts))
    ]

    valid_alloc = _approximate_stratified_allocation(remaining_counts, n_valid, seed=seed + 1)

    alloc_rows = [
        (float(labels[i]), int(test_alloc[i]), int(valid_alloc[i]))
        for i in range(len(labels))
    ]

    alloc_schema = StructType([
        StructField(label_col, DoubleType(), nullable=False),
        StructField("_n_test", LongType(), nullable=False),
        StructField("_n_valid", LongType(), nullable=False),
    ])

    alloc_df = spark_session.createDataFrame(alloc_rows, alloc_schema)

    # Random ổn định bằng hash thay vì orderBy(rand()).
    w = Window.partitionBy(label_col).orderBy(
        F.xxhash64(F.col("_split_row_id"), F.lit(int(seed))).asc(),
        F.col("_split_row_id").asc(),
    )

    split_df = (
        base_df
        .join(F.broadcast(alloc_df), on=label_col, how="left")
        .withColumn("_rn", F.row_number().over(w))
        .withColumn(
            "_split",
            F.when(F.col("_rn") <= F.col("_n_test"), F.lit("test"))
             .when(F.col("_rn") <= F.col("_n_test") + F.col("_n_valid"), F.lit("valid"))
             .otherwise(F.lit("train"))
        )
        .drop("_n_test", "_n_valid", "_rn", "_split_row_id")
    )

    if cache_result:
        split_df = split_df.persist(StorageLevel.MEMORY_AND_DISK)
        split_df.count()

    train_df = split_df.filter(F.col("_split") == "train").drop("_split")
    valid_df = split_df.filter(F.col("_split") == "valid").drop("_split")
    test_df = split_df.filter(F.col("_split") == "test").drop("_split")

    if return_split_df:
        return train_df, valid_df, test_df, split_df

    return train_df, valid_df, test_df


def print_split_report_spark(train_df, valid_df, test_df, label_col="label"):
    """In số dòng và tỷ lệ label=1 cho từng split."""
    for name, part in [("Train", train_df), ("Valid", valid_df), ("Test ", test_df)]:
        n, pr = get_count_and_positive_rate(part, label_col=label_col)
        print(f"{name}: rows={n:,} | columns={len(part.columns)} | positive_rate={pr:.6f}")
        part.groupBy(label_col).count().orderBy(label_col).show()
```

## 11. Feature pipeline

Cell này làm 3 việc:

1. Numeric columns → cast sang `double`.
2. String categorical columns → `StringIndexer`.
3. Ghép tất cả feature vào cột vector `features` để SynapseML LightGBM sử dụng.


```python
class SparkTaxiFeaturePipeline:
    """
    Feature pipeline đơn giản cho Spark/SynapseML LightGBM.

    Thiết kế:
    - Không tạo thêm __isnull features để giữ tương đồng với notebook LightGBM thường.
    - Numeric null/NaN được giữ trong vector; LightGBM có thể xử lý missing value.
    - String categorical dùng StringIndexer(handleInvalid='keep').
    - Numeric categorical được cast double và khai báo categoricalSlotIndexes cho SynapseML.
    """

    def __init__(self, categorical_cols, numeric_cols, exclude_cols, label_col="label"):
        self.categorical_cols = list(categorical_cols)
        self.numeric_cols = list(numeric_cols)
        self.exclude_cols = set(exclude_cols)
        self.label_col = label_col

        self.numeric_feature_cols_ = []
        self.categorical_numeric_cols_ = []
        self.categorical_string_cols_ = []
        self.string_indexer_models_ = []
        self.string_indexed_cols_ = []
        self.feature_cols_ = []
        self.categorical_feature_cols_ = []
        self.categorical_slot_indexes_ = []
        self.assembler_ = None

    def fit(self, train_df):
        """Fit StringIndexer trên train set và xác định thứ tự feature."""
        available = set(train_df.columns)
        schema_map = {f.name: f.dataType for f in train_df.schema.fields}

        self.numeric_feature_cols_ = [
            c for c in self.numeric_cols
            if c in available and c not in self.exclude_cols
        ]

        self.categorical_string_cols_ = []
        self.categorical_numeric_cols_ = []

        for c in self.categorical_cols:
            if c not in available or c in self.exclude_cols:
                continue

            dtype = schema_map[c]
            if isinstance(dtype, T.StringType):
                self.categorical_string_cols_.append(c)
            else:
                self.categorical_numeric_cols_.append(c)

        df = train_df
        self.string_indexer_models_ = []
        self.string_indexed_cols_ = []

        for c in self.categorical_string_cols_:
            idx_col = f"{c}__idx"
            indexer = StringIndexer(inputCol=c, outputCol=idx_col, handleInvalid="keep")
            model = indexer.fit(df)
            self.string_indexer_models_.append(model)
            self.string_indexed_cols_.append(idx_col)
            df = model.transform(df)

        self.feature_cols_ = (
            self.numeric_feature_cols_
            + self.categorical_numeric_cols_
            + self.string_indexed_cols_
        )

        self.categorical_feature_cols_ = self.categorical_numeric_cols_ + self.string_indexed_cols_
        self.categorical_slot_indexes_ = [
            self.feature_cols_.index(c)
            for c in self.categorical_feature_cols_
        ]

        self.assembler_ = VectorAssembler(
            inputCols=self.feature_cols_,
            outputCol="features",
            handleInvalid="keep",
        )

        return self

    def transform(self, input_df):
        """Transform input DataFrame thành DataFrame có cột features."""
        df = input_df

        # Cast numeric và categorical numeric sang double.
        for c in self.numeric_feature_cols_ + self.categorical_numeric_cols_:
            if c in df.columns:
                df = df.withColumn(c, F.col(c).cast("double"))
                df = df.withColumn(
                    c,
                    F.when(
                        F.isnan(F.col(c))
                        | (F.col(c) == float("inf"))
                        | (F.col(c) == float("-inf")),
                        F.lit(None).cast("double"),
                    ).otherwise(F.col(c))
                )

        # Áp dụng StringIndexer models đã fit từ train set.
        for model in self.string_indexer_models_:
            df = model.transform(df)

        # Ghép feature columns thành vector.
        df = self.assembler_.transform(df)

        return df
```

## 13. Missing summary before vector assembly

Cell này giúp bạn kiểm tra các cột feature còn null/missing bao nhiêu trước khi train.


```python
def build_missing_summary_spark(input_df, feature_cols):
    """Tính null/missing count cho từng feature column."""
    total_rows = input_df.count()
    schema_map = {f.name: f.dataType for f in input_df.schema.fields}

    exprs = []
    for c in feature_cols:
        if c not in input_df.columns:
            continue

        dtype = schema_map[c]
        if isinstance(dtype, (T.DoubleType, T.FloatType, T.IntegerType, T.LongType, T.ShortType, T.ByteType, T.DecimalType)):
            missing_expr = F.sum(
                F.when(F.col(c).isNull() | F.isnan(F.col(c).cast("double")), 1).otherwise(0)
            ).alias(c)
        else:
            missing_expr = F.sum(F.when(F.col(c).isNull(), 1).otherwise(0)).alias(c)

        exprs.append(missing_expr)

    row = input_df.select(exprs).first().asDict()

    result = pd.DataFrame([
        {
            "column": c,
            "missing_count": int(row.get(c, 0)),
            "missing_rate_pct": (int(row.get(c, 0)) / total_rows * 100) if total_rows > 0 else 0.0,
        }
        for c in row.keys()
    ]).sort_values("missing_count", ascending=False)

    return result


raw_feature_cols = CATEGORICAL_COLS + NUMERIC_COLS
missing_summary_df = build_missing_summary_spark(work_df, raw_feature_cols)
display(missing_summary_df)

missing_path = f"{OUTPUT_DIR}/feature_missing_summary.csv"
missing_summary_df.to_csv(missing_path, index=False)
print("Saved:", missing_path)
```

## 14. SynapseML LightGBM helper functions

Các hàm dưới đây giúp:

- Gộp train + valid cho SynapseML thông qua `validationIndicatorCol`.
- Chuẩn hóa params để tránh nhầm tên param giữa LightGBM Python và SynapseML.
- Train/evaluate AUC.


```python
def build_lgbm_train_df(train_df, valid_df, num_partitions=None, persist=True):
    """
    Gộp train + valid cho SynapseML LightGBM.

    SynapseML dùng validationIndicatorCol:
    - is_validation = False: train rows
    - is_validation = True : validation rows
    """
    if num_partitions is None:
        num_partitions = LGBM_NUM_PARTITIONS

    train_part = (
        train_df
        .select(F.col("features"), F.col("label").cast("double").alias("label"))
        .withColumn("is_validation", F.lit(False))
    )

    valid_part = (
        valid_df
        .select(F.col("features"), F.col("label").cast("double").alias("label"))
        .withColumn("is_validation", F.lit(True))
    )

    out_df = train_part.unionByName(valid_part)

    if num_partitions is not None:
        out_df = out_df.repartition(int(num_partitions))

    if persist:
        out_df = out_df.persist(StorageLevel.MEMORY_AND_DISK)
        out_df.count()

    return out_df


SYNAPSEML_TUNABLE_PARAM_NAMES = {
    "learningRate", "numLeaves", "maxDepth", "minDataInLeaf",
    "baggingFraction", "baggingFreq", "featureFraction",
    "lambdaL1", "lambdaL2", "minGainToSplit", "maxBin",
}

SYNAPSEML_FIXED_PARAM_NAMES = {
    "numIterations", "earlyStoppingRound", "numTasks", "numThreads",
    "maxStreamingOMPThreads", "timeout", "useBarrierExecutionMode",
    "useSingleDatasetMode", "dataTransferMode", "verbosity",
    "driverListenPort", "defaultListenPort",
}


def _to_bool(value):
    """Convert bool/int/string thành bool rõ ràng."""
    if isinstance(value, bool):
        return value
    if isinstance(value, int):
        return bool(value)
    if isinstance(value, str):
        v = value.strip().lower()
        if v in {"true", "1", "yes", "y"}:
            return True
        if v in {"false", "0", "no", "n"}:
            return False
    raise ValueError(f"Không thể convert sang bool: {value}")


def validate_synapseml_params(params):
    """Validate params để Optuna không sinh cấu hình sai hoặc quá bất hợp lý."""
    p = dict(params)

    if not (0.0 < float(p["learningRate"]) <= 1.0):
        raise ValueError(f"learningRate không hợp lệ: {p['learningRate']}")
    if int(p["numLeaves"]) < 2:
        raise ValueError(f"numLeaves phải >= 2: {p['numLeaves']}")
    if int(p["maxDepth"]) != -1 and int(p["maxDepth"]) < 1:
        raise ValueError(f"maxDepth phải = -1 hoặc >= 1: {p['maxDepth']}")
    if int(p["minDataInLeaf"]) < 1:
        raise ValueError(f"minDataInLeaf phải >= 1: {p['minDataInLeaf']}")
    if not (0.0 < float(p["baggingFraction"]) <= 1.0):
        raise ValueError(f"baggingFraction không hợp lệ: {p['baggingFraction']}")
    if int(p["baggingFreq"]) < 0:
        raise ValueError(f"baggingFreq phải >= 0: {p['baggingFreq']}")
    if not (0.0 < float(p["featureFraction"]) <= 1.0):
        raise ValueError(f"featureFraction không hợp lệ: {p['featureFraction']}")
    if float(p["lambdaL1"]) < 0 or float(p["lambdaL2"]) < 0:
        raise ValueError("lambdaL1/lambdaL2 phải >= 0")
    if float(p["minGainToSplit"]) < 0:
        raise ValueError(f"minGainToSplit phải >= 0: {p['minGainToSplit']}")
    if int(p["maxBin"]) < 2:
        raise ValueError(f"maxBin phải >= 2: {p['maxBin']}")
    if int(p["numIterations"]) < 1:
        raise ValueError(f"numIterations phải >= 1: {p['numIterations']}")
    if int(p["earlyStoppingRound"]) < 1:
        raise ValueError(f"earlyStoppingRound phải >= 1: {p['earlyStoppingRound']}")
    if int(p["numTasks"]) < 1 or int(p["numThreads"]) < 1:
        raise ValueError("numTasks/numThreads phải >= 1")

    return p


def with_fixed_synapseml_params(params):
    """
    Nhận params tunable rồi thêm fixed config cho SynapseML.

    Quan trọng:
    - Chỉ dùng tên params của SynapseML như learningRate, numLeaves...
    - Không dùng tên LightGBM Python như learning_rate, num_leaves để tránh nhầm.
    """
    params = dict(params)

    allowed = SYNAPSEML_TUNABLE_PARAM_NAMES | SYNAPSEML_FIXED_PARAM_NAMES
    unknown_keys = sorted(set(params) - allowed)
    if unknown_keys:
        raise ValueError(f"Unknown SynapseML param keys: {unknown_keys}")

    out = dict(params)

    # Fixed training controls.
    out["numIterations"] = int(out.get("numIterations", N_ESTIMATORS_MAX))
    out["earlyStoppingRound"] = int(out.get("earlyStoppingRound", EARLY_STOPPING_ROUNDS))

    # Distributed LightGBM controls.
    out["numTasks"] = int(out.get("numTasks", LGBM_NUM_TASKS))
    out["numThreads"] = int(out.get("numThreads", LGBM_NUM_THREADS))
    out["maxStreamingOMPThreads"] = int(out.get("maxStreamingOMPThreads", LGBM_NUM_THREADS))

    # Network/transfer controls.
    out["timeout"] = int(out.get("timeout", 600))
    out["useBarrierExecutionMode"] = _to_bool(out.get("useBarrierExecutionMode", True))
    out["useSingleDatasetMode"] = _to_bool(out.get("useSingleDatasetMode", True))
    out["dataTransferMode"] = out.get("dataTransferMode", "streaming")
    out["verbosity"] = int(out.get("verbosity", -1))

    # Tunable model params with defaults.
    out["learningRate"] = float(out.get("learningRate", 0.05))
    out["numLeaves"] = int(out.get("numLeaves", 31))
    out["maxDepth"] = int(out.get("maxDepth", -1))
    out["minDataInLeaf"] = int(out.get("minDataInLeaf", 20))
    out["baggingFraction"] = float(out.get("baggingFraction", 1.0))
    out["baggingFreq"] = int(out.get("baggingFreq", 0))
    out["featureFraction"] = float(out.get("featureFraction", 1.0))
    out["lambdaL1"] = float(out.get("lambdaL1", 0.0))
    out["lambdaL2"] = float(out.get("lambdaL2", 0.0))
    out["minGainToSplit"] = float(out.get("minGainToSplit", 0.0))
    out["maxBin"] = int(out.get("maxBin", 255))

    if USE_FIXED_PORTS:
        out["driverListenPort"] = int(out.get("driverListenPort", 12400))
        out["defaultListenPort"] = int(out.get("defaultListenPort", 12401))
    else:
        out.pop("driverListenPort", None)
        out.pop("defaultListenPort", None)

    return validate_synapseml_params(out)


def build_synapseml_lgbm_classifier(
    synapseml_params,
    categorical_slot_indexes=None,
    shap_col=None,
    validation_col="is_validation",
):
    """Tạo LightGBMClassifier của SynapseML từ params đã chuẩn hóa."""
    p = with_fixed_synapseml_params(synapseml_params)

    kwargs = {
        "objective": "binary",
        "metric": "auc",
        "featuresCol": "features",
        "labelCol": "label",
        "validationIndicatorCol": validation_col,
        "learningRate": float(p["learningRate"]),
        "numLeaves": int(p["numLeaves"]),
        "maxDepth": int(p["maxDepth"]),
        "minDataInLeaf": int(p["minDataInLeaf"]),
        "baggingFraction": float(p["baggingFraction"]),
        "baggingFreq": int(p["baggingFreq"]),
        "featureFraction": float(p["featureFraction"]),
        "lambdaL1": float(p["lambdaL1"]),
        "lambdaL2": float(p["lambdaL2"]),
        "minGainToSplit": float(p["minGainToSplit"]),
        "maxBin": int(p["maxBin"]),
        "numIterations": int(p["numIterations"]),
        "earlyStoppingRound": int(p["earlyStoppingRound"]),
        "numTasks": int(p["numTasks"]),
        "numThreads": int(p["numThreads"]),
        "maxStreamingOMPThreads": int(p["maxStreamingOMPThreads"]),
        "timeout": int(p["timeout"]),
        "useBarrierExecutionMode": bool(p["useBarrierExecutionMode"]),
        "useSingleDatasetMode": bool(p["useSingleDatasetMode"]),
        "dataTransferMode": p["dataTransferMode"],
        "verbosity": int(p["verbosity"]),
    }

    if USE_FIXED_PORTS:
        kwargs["driverListenPort"] = int(p["driverListenPort"])
        kwargs["defaultListenPort"] = int(p["defaultListenPort"])

    if categorical_slot_indexes:
        kwargs["categoricalSlotIndexes"] = [int(x) for x in categorical_slot_indexes]

    # Chỉ bật SHAP khi thật sự cần vì transform sẽ chậm hơn nhiều.
    if shap_col:
        kwargs["featuresShapCol"] = shap_col

    return LightGBMClassifier(**kwargs)


_auc_evaluator = BinaryClassificationEvaluator(
    labelCol="label",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderROC",
)


def evaluate_auc(model, eval_df, return_predictions=False, persist_predictions=False):
    """
    Evaluate AUC.

    - Optuna/baseline: return_predictions=False để tiết kiệm memory.
    - Final model: return_predictions=True để tính classification report / SHAP.
    """
    pred_df = model.transform(eval_df)
    auc = float(_auc_evaluator.evaluate(pred_df))

    if not return_predictions:
        return auc, None

    keep_cols = [
        "features", "label", "rawPrediction", "probability", "prediction", "featuresShap"
    ]
    keep_cols = [c for c in keep_cols if c in pred_df.columns]
    pred_df = pred_df.select(*keep_cols)

    if persist_predictions:
        pred_df = pred_df.persist(StorageLevel.MEMORY_AND_DISK)
        pred_df.count()

    return auc, pred_df


def extract_probability_label_df(pred_df, limit=None):
    """Collect label và probability(label=1) về pandas để tính metric ngoài Spark."""
    score_df = pred_df.select(
        F.col("label").cast("double").alias("y_true"),
        vector_to_array("probability")[1].cast("double").alias("pred_score"),
    )

    if limit is not None:
        score_df = score_df.limit(int(limit))

    return score_df.toPandas()


def safe_get_best_iteration(model):
    """Lấy best iteration nếu SynapseML model expose method tương ứng."""
    for attr in ["getBestIteration", "bestIteration", "best_iteration"]:
        try:
            value = getattr(model, attr)
            value = value() if callable(value) else value
            if value is not None:
                return int(value)
        except Exception:
            pass
    return None
```

## 15. Build reusable LightGBM train/eval DataFrames

DataFrame `lgbm_train_df` gồm cả train và validation, có cột `is_validation`. SynapseML LightGBM dùng cột này để early stopping.


```python
lgbm_train_df = build_lgbm_train_df(
    train_df=train_features_df,
    valid_df=valid_features_df,
    num_partitions=LGBM_NUM_PARTITIONS,
    persist=True,
)

valid_eval_df = valid_features_df.select("features", "label")
test_eval_df = test_features_df.select("features", "label")

n_lgbm_train = lgbm_train_df.count()
n_valid_eval = valid_eval_df.count()
n_test_eval = test_eval_df.count()

print("LGBM train + valid rows:", n_lgbm_train)
print("Valid rows             :", n_valid_eval)
print("Test rows              :", n_test_eval)
```

## 16. Baseline SynapseML LightGBM

Baseline dùng params gần với LightGBM thường để có mốc so sánh với Optuna.


```python
BASELINE_TUNABLE_PARAMS = {
    "learningRate": 0.05,
    "numLeaves": 31,
    "maxDepth": -1,
    "minDataInLeaf": 20,
    "baggingFraction": 1.0,
    "baggingFreq": 0,
    "featureFraction": 1.0,
    "lambdaL1": 1e-8,
    "lambdaL2": 1e-8,
    "minGainToSplit": 0.0,
    # 255 khớp LightGBM thường. Nếu cluster hay lỗi native memory, đổi về 127.
    "maxBin": 255,
}


def train_lgbm_with_params(
    tunable_params,
    shap_col=None,
    return_test_predictions=False,
    persist_test_predictions=False,
):
    """
    Train SynapseML LightGBM trên lgbm_train_df đã cache và evaluate valid/test AUC.
    """
    full_params = with_fixed_synapseml_params(tunable_params)

    estimator = build_synapseml_lgbm_classifier(
        synapseml_params=full_params,
        categorical_slot_indexes=feature_pipeline.categorical_slot_indexes_,
        shap_col=shap_col,
        validation_col="is_validation",
    )

    start = time.time()
    model = estimator.fit(lgbm_train_df)
    fit_seconds = time.time() - start

    valid_auc, _ = evaluate_auc(model, valid_eval_df, return_predictions=False)
    test_auc, test_pred_df = evaluate_auc(
        model,
        test_eval_df,
        return_predictions=return_test_predictions,
        persist_predictions=persist_test_predictions,
    )

    result = {
        "valid_auc": float(valid_auc),
        "test_auc": float(test_auc),
        "best_iteration": safe_get_best_iteration(model),
        "fit_seconds": float(fit_seconds),
    }

    return result, model, test_pred_df


baseline_result, baseline_model, _ = train_lgbm_with_params(
    BASELINE_TUNABLE_PARAMS,
    shap_col=None,
    return_test_predictions=False,
)

baseline_valid_auc = baseline_result["valid_auc"]
baseline_test_auc = baseline_result["test_auc"]

print("Baseline result:")
print(json.dumps(baseline_result, indent=2, default=str))

gc.collect()
```

## 17. Optuna tuning

Optuna chỉ tối ưu trên validation AUC. Không bật SHAP trong trial để tránh mỗi trial bị chậm quá nhiều.


```python
def suggest_synapseml_params(trial):
    """Search space cho SynapseML LightGBM."""
    params = {
        # Learning speed.
        "learningRate": trial.suggest_float("learningRate", 0.01, 0.20, log=True),

        # Tree complexity.
        "numLeaves": trial.suggest_categorical("numLeaves", [15, 31, 63, 127]),
        "maxDepth": trial.suggest_categorical("maxDepth", [-1, 4, 6, 8, 10, 12]),
        "minDataInLeaf": trial.suggest_categorical("minDataInLeaf", [10, 20, 50, 100, 200]),

        # Row sampling.
        "baggingFraction": trial.suggest_float("baggingFraction", 0.7, 1.0),
        "baggingFreq": trial.suggest_categorical("baggingFreq", [0, 1, 3, 5]),

        # Feature sampling.
        "featureFraction": trial.suggest_float("featureFraction", 0.7, 1.0),

        # Regularization.
        "lambdaL1": trial.suggest_float("lambdaL1", 1e-8, 10.0, log=True),
        "lambdaL2": trial.suggest_float("lambdaL2", 1e-8, 10.0, log=True),
        "minGainToSplit": trial.suggest_float("minGainToSplit", 0.0, 1.0),

        # Histogram bins.
        "maxBin": trial.suggest_categorical("maxBin", [63, 127, 255]),
    }

    return with_fixed_synapseml_params(params)


def objective(trial):
    """Optuna objective: maximize validation AUC."""
    params = suggest_synapseml_params(trial)

    try:
        result, model, _ = train_lgbm_with_params(
            params,
            shap_col=None,
            return_test_predictions=False,
        )

        trial.set_user_attr("test_auc", result["test_auc"])
        trial.set_user_attr("best_iteration", result["best_iteration"])
        trial.set_user_attr("fit_seconds", result["fit_seconds"])

        return float(result["valid_auc"])

    except Exception as e:
        print(f"[Trial {trial.number}] failed:", repr(e))
        if DEBUG_TRIAL_ERRORS:
            raise
        return 0.5

    finally:
        gc.collect()


study = optuna.create_study(
    direction="maximize",
    sampler=TPESampler(seed=RANDOM_STATE),
    pruner=MedianPruner(n_startup_trials=10, n_warmup_steps=0),
)

# Enqueue baseline để Optuna luôn có mốc baseline trong study.
study.enqueue_trial(BASELINE_TUNABLE_PARAMS)

if RUN_OPTUNA:
    study.optimize(objective, n_trials=N_TRIALS, show_progress_bar=True)
else:
    # Nếu không muốn tuning lâu, chỉ chạy baseline trial đã enqueue.
    study.optimize(objective, n_trials=1, show_progress_bar=True)

print("Best validation AUC:", study.best_value)
print("Best params:")
print(study.best_params)
```

```python
trials_df = study.trials_dataframe()
trials_path = f"{OUTPUT_DIR}/optuna_synapseml_lgbm_trials.csv"
trials_df.to_csv(trials_path, index=False)

print("Saved:", trials_path)
display(trials_df.sort_values("value", ascending=False).head(10))

try:
    from optuna.importance import get_param_importances

    param_importance = get_param_importances(study)
    param_importance_df = pd.DataFrame({
        "parameter": list(param_importance.keys()),
        "importance": list(param_importance.values()),
    }).sort_values("importance", ascending=False)

    param_importance_path = f"{OUTPUT_DIR}/optuna_synapseml_param_importance.csv"
    param_importance_df.to_csv(param_importance_path, index=False)

    print("Saved:", param_importance_path)
    display(param_importance_df)
except Exception as e:
    print("Param importance skipped:", repr(e))
```

## 20. Feature importance and optional SHAP

- Native feature importance nhẹ hơn SHAP.
- SHAP trong SynapseML nặng hơn nhiều, nên chỉ bật bằng `RUN_FINAL_SHAP=True`.


```python
def get_synapseml_feature_importance_df(model, feature_cols):
    """Đọc feature importance từ SynapseML model nếu version hiện tại hỗ trợ."""
    importances = None

    # SynapseML versions khác nhau có thể expose importance khác nhau.
    for expr in [
        "model.getFeatureImportances()",
        "model.getLightGBMBooster().getFeatureImportances()",
    ]:
        try:
            importances = eval(expr)
            break
        except Exception:
            pass

    if importances is None:
        print("Could not read native feature_importances from SynapseML model.")
        return pd.DataFrame({"feature": feature_cols, "importance": np.nan})

    importances = list(importances)
    n = min(len(importances), len(feature_cols))

    if len(importances) != len(feature_cols):
        print("Warning: importance length mismatch:", len(importances), len(feature_cols))

    return pd.DataFrame({
        "feature": feature_cols[:n],
        "importance": importances[:n],
    }).sort_values("importance", ascending=False)


importance_df = get_synapseml_feature_importance_df(final_model, feature_pipeline.feature_cols_)
importance_path = f"{OUTPUT_DIR}/final_synapseml_feature_importance.csv"
importance_df.to_csv(importance_path, index=False)

print("Saved:", importance_path)
display(importance_df.head(30))
```

```python
import matplotlib.pyplot as plt

plot_df = importance_df.dropna().head(20).sort_values("importance")

if len(plot_df) > 0:
    plt.figure(figsize=(8, 6))
    plt.barh(plot_df["feature"], plot_df["importance"])
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    plt.title("Top SynapseML LightGBM Feature Importance")
    plt.show()
else:
    print("Native feature importance not available or empty.")
```

```python
def compute_top_shap_from_predictions(pred_df, feature_cols, shap_col="featuresShap", top_n=20, sample_size=None, seed=42):
    """Tính mean(|SHAP|) cho từng feature từ prediction DataFrame."""
    if shap_col not in pred_df.columns:
        raise ValueError(
            f"Column {shap_col!r} không tồn tại. "
            "Hãy bật RUN_FINAL_SHAP=True hoặc estimator phải có featuresShapCol."
        )

    shap_source_df = pred_df

    if sample_size is not None:
        total = pred_df.count()
        if total > sample_size:
            frac = min(1.0, float(sample_size) / float(total) * 1.5)
            shap_source_df = pred_df.sample(False, frac, seed=seed).limit(int(sample_size))

    shap_df = shap_source_df.withColumn("shap_array", vector_to_array(shap_col))
    shap_size = shap_df.select(F.size("shap_array").alias("shap_size")).first()["shap_size"]

    shap_long_df = shap_df.select(
        F.posexplode("shap_array").alias("feature_index", "shap_value")
    )

    # Một số version có bias/base value ở cuối vector SHAP. Nếu có, bỏ phần đó.
    if shap_size == len(feature_cols) + 1:
        shap_long_df = shap_long_df.filter(F.col("feature_index") < len(feature_cols))

    feature_mapping_df = spark.createDataFrame(
        [(i, name) for i, name in enumerate(feature_cols)],
        ["feature_index", "feature"],
    )

    shap_all_df = (
        shap_long_df
        .groupBy("feature_index")
        .agg(F.avg(F.abs("shap_value")).alias("mean_abs_shap"))
        .join(F.broadcast(feature_mapping_df), on="feature_index", how="left")
        .orderBy(F.desc("mean_abs_shap"))
    )

    return shap_all_df.limit(top_n), shap_all_df


if RUN_FINAL_SHAP and final_test_pred_df is not None and "featuresShap" in final_test_pred_df.columns:
    shap_top_df_spark, shap_all_df_spark = compute_top_shap_from_predictions(
        pred_df=final_test_pred_df,
        feature_cols=feature_pipeline.feature_cols_,
        shap_col="featuresShap",
        top_n=20,
        sample_size=SHAP_SAMPLE_SIZE,
        seed=RANDOM_STATE,
    )

    shap_top_pdf = shap_top_df_spark.toPandas()
    shap_path = f"{OUTPUT_DIR}/final_synapseml_shap_top_features.csv"
    shap_top_pdf.to_csv(shap_path, index=False)

    print("Saved:", shap_path)
    display(shap_top_pdf)
else:
    shap_top_pdf = pd.DataFrame(columns=["feature_index", "mean_abs_shap", "feature"])
    print("Final SHAP skipped. Set RUN_FINAL_SHAP=True to compute SynapseML featuresShap.")
```

```python
if len(shap_top_pdf) > 0:
    plot_df = shap_top_pdf.sort_values("mean_abs_shap")

    plt.figure(figsize=(8, 6))
    plt.barh(plot_df["feature"], plot_df["mean_abs_shap"])
    plt.xlabel("Mean absolute SHAP value")
    plt.ylabel("Feature")
    plt.title("Top SynapseML SHAP Features")
    plt.show()
else:
    print("No SHAP plot because SHAP was skipped or unavailable.")
```

## 21. Optional: Synthetic anomaly experiment

Phần này chỉ chạy khi `RUN_SYNTHETIC_EXPERIMENTS=True`. Mục tiêu là kiểm tra detector có nhạy hơn khi ta cố tình inject anomaly vào target batch hay không.


```python
ANOMALY_LEVELS = [0.0, 0.05, 0.10, 0.20, 0.30, 0.50, 0.75]
ANOMALY_RANDOM_STATE = 2025


def inject_synthetic_anomalies_spark(target_df, anomaly_fraction=0.30, random_state=42):
    """
    Inject anomaly đơn giản vào một phần target rows.

    Đây chỉ là synthetic test, không dùng để clean dữ liệu thật.
    """
    data = (
        target_df
        .withColumn("_r_anom", F.rand(int(random_state)))
        .withColumn("synthetic_anomaly", (F.col("_r_anom") < float(anomaly_fraction)).cast("int"))
        .withColumn("_anom_group", (F.rand(int(random_state) + 1) * 5).cast("int"))
    )

    is_anom = F.col("synthetic_anomaly") == 1

    if "trip_distance" in data.columns:
        data = data.withColumn(
            "trip_distance",
            F.when(is_anom & (F.col("_anom_group") == 0), F.col("trip_distance") * F.lit(5.0))
             .otherwise(F.col("trip_distance"))
        )

    if "total_amount" in data.columns:
        data = data.withColumn(
            "total_amount",
            F.when(is_anom & (F.col("_anom_group") == 1), F.col("total_amount") * F.lit(4.0))
             .otherwise(F.col("total_amount"))
        )

    if "fare_amount" in data.columns:
        data = data.withColumn(
            "fare_amount",
            F.when(is_anom & (F.col("_anom_group") == 2), F.col("fare_amount") * F.lit(4.0))
             .otherwise(F.col("fare_amount"))
        )

    if "store_and_fwd_flag" in data.columns:
        data = data.withColumn(
            "store_and_fwd_flag",
            F.when(is_anom & (F.col("_anom_group") == 3), F.lit("ANOM"))
             .otherwise(F.col("store_and_fwd_flag"))
        )

    for loc_col in ["pulocationid", "dolocationid"]:
        if loc_col in data.columns:
            data = data.withColumn(
                loc_col,
                F.when(is_anom & (F.col("_anom_group") == 4), F.lit(999))
                 .otherwise(F.col(loc_col))
            )

    return data.drop("_r_anom", "_anom_group")


def train_synapseml_on_work_df(batch_work_df, synapseml_params, random_state=42, shap_col=None, return_artifacts=False):
    """
    Train/evaluate trọn gói trên một work_df bất kỳ.

    Dùng cho:
    - Synthetic experiments.
    - Batch monitoring loop.
    """
    local_split_df = None
    local_lgbm_train_df = None
    local_train_features_df = None
    local_valid_features_df = None
    local_test_features_df = None
    local_test_pred_df = None

    try:
        local_train_raw_df, local_valid_raw_df, local_test_raw_df, local_split_df = stratified_split_sklearn_like_spark(
            input_df=batch_work_df,
            label_col="label",
            test_size=TEST_SIZE,
            valid_size=VALID_SIZE,
            seed=random_state,
            cache_result=True,
            return_split_df=True,
        )

        local_pipeline = SparkTaxiFeaturePipeline(
            categorical_cols=CATEGORICAL_COLS,
            numeric_cols=NUMERIC_COLS,
            exclude_cols=EXCLUDE_COLS,
            label_col="label",
        ).fit(local_train_raw_df)

        local_train_features_df = local_pipeline.transform(local_train_raw_df).select("features", "label").persist(StorageLevel.MEMORY_AND_DISK)
        local_valid_features_df = local_pipeline.transform(local_valid_raw_df).select("features", "label").persist(StorageLevel.MEMORY_AND_DISK)
        local_test_features_df = local_pipeline.transform(local_test_raw_df).select("features", "label").persist(StorageLevel.MEMORY_AND_DISK)

        local_train_features_df.count()
        local_valid_features_df.count()
        local_test_features_df.count()

        local_lgbm_train_df = build_lgbm_train_df(
            train_df=local_train_features_df,
            valid_df=local_valid_features_df,
            num_partitions=LGBM_NUM_PARTITIONS,
            persist=True,
        )

        estimator = build_synapseml_lgbm_classifier(
            synapseml_params=synapseml_params,
            categorical_slot_indexes=local_pipeline.categorical_slot_indexes_,
            shap_col=shap_col,
            validation_col="is_validation",
        )

        start = time.time()
        model = estimator.fit(local_lgbm_train_df)
        fit_seconds = time.time() - start

        valid_auc, _ = evaluate_auc(model, local_valid_features_df.select("features", "label"), return_predictions=False)
        test_auc, local_test_pred_df = evaluate_auc(
            model,
            local_test_features_df.select("features", "label"),
            return_predictions=return_artifacts,
            persist_predictions=return_artifacts,
        )

        result = {
            "valid_auc": float(valid_auc),
            "test_auc": float(test_auc),
            "best_iteration": safe_get_best_iteration(model),
            "fit_seconds": float(fit_seconds),
            "feature_count": len(local_pipeline.feature_cols_),
        }

        artifacts = None
        if return_artifacts:
            artifacts = {
                "model": model,
                "pipeline": local_pipeline,
                "test_pred_df": local_test_pred_df,
                "feature_cols": local_pipeline.feature_cols_,
            }

        return result, artifacts

    finally:
        if not return_artifacts:
            for obj in [
                local_test_pred_df,
                local_lgbm_train_df,
                local_train_features_df,
                local_valid_features_df,
                local_test_features_df,
                local_split_df,
            ]:
                try:
                    if obj is not None:
                        obj.unpersist()
                except Exception:
                    pass
        gc.collect()


if RUN_SYNTHETIC_EXPERIMENTS:
    synthetic_results = []

    for frac in ANOMALY_LEVELS:
        print("Running synthetic anomaly fraction:", frac)

        temp_positive = inject_synthetic_anomalies_spark(
            positive_df,
            anomaly_fraction=frac,
            random_state=ANOMALY_RANDOM_STATE,
        )

        temp_negative = negative_df.withColumn("synthetic_anomaly", F.lit(0))

        temp_work_df = (
            temp_positive
            .unionByName(temp_negative, allowMissingColumns=True)
            .repartition(SPARK_SQL_SHUFFLE_PARTITIONS)
            .persist(StorageLevel.MEMORY_AND_DISK)
        )
        temp_work_df.count()

        try:
            result, _ = train_synapseml_on_work_df(
                temp_work_df,
                synapseml_params=selected_synapseml_params,
                random_state=RANDOM_STATE,
                shap_col=None,
                return_artifacts=False,
            )

            result["anomaly_fraction"] = float(frac)
            result["n_rows"] = int(temp_work_df.count())
            result["n_synthetic_anomaly"] = int(temp_work_df.filter(F.col("synthetic_anomaly") == 1).count())
            synthetic_results.append(result)

            print(result)
        finally:
            temp_work_df.unpersist()

    synthetic_df = pd.DataFrame(synthetic_results)
    synthetic_path = f"{OUTPUT_DIR}/synthetic_anomaly_sweep.csv"
    synthetic_df.to_csv(synthetic_path, index=False)

    print("Saved:", synthetic_path)
    display(synthetic_df)
else:
    print("Synthetic experiments skipped. Set RUN_SYNTHETIC_EXPERIMENTS=True to run.")
```

## 22. Optional: Batch AUC monitoring

Phần này chạy nhiều target batches liên tiếp để tạo chuỗi AUC. Chỉ chạy khi `RUN_BATCH_MONITORING=True`.


```python
def build_dataset_for_target_batch_spark(
    source_df,
    target_processing_hour,
    min_rows_per_group=MIN_ROWS_PER_GROUP,
    max_rows_per_group=TARGET_SAMPLE_PER_GROUP,
    random_state=RANDOM_STATE,
):
    """Build work_df cho một target batch cụ thể."""
    positive = sample_by_processing_hour_spark(
        source_df=source_df,
        processing_hour=target_processing_hour,
        label=1,
        max_rows=max_rows_per_group,
        random_state=random_state,
        oversample_factor=OVERSAMPLE_FACTOR,
    )

    positive_rows = positive.count()
    if positive_rows < min_rows_per_group:
        return None, {
            "status": "skipped",
            "reason": "positive_rows_too_few",
            "target_processing_date_hour": str(target_processing_hour),
            "positive_rows": int(positive_rows),
        }

    comparison_hours = get_comparison_processing_hours(target_processing_hour)

    negative_parts = []
    negative_info = []

    for i, compare_hour in enumerate(comparison_hours):
        part = sample_by_processing_hour_spark(
            source_df=source_df,
            processing_hour=compare_hour,
            label=0,
            max_rows=max_rows_per_group,
            random_state=random_state + i + 100,
            oversample_factor=OVERSAMPLE_FACTOR,
        )

        rows = part.count()
        negative_info.append({"comparison_hour": str(compare_hour), "rows": int(rows)})

        if rows >= min_rows_per_group:
            negative_parts.append(part)

    if len(negative_parts) == 0:
        return None, {
            "status": "skipped",
            "reason": "negative_rows_too_few",
            "target_processing_date_hour": str(target_processing_hour),
            "positive_rows": int(positive_rows),
            "negative_info": negative_info,
        }

    negative = negative_parts[0]
    for p in negative_parts[1:]:
        negative = negative.unionByName(p, allowMissingColumns=True)

    work = positive.unionByName(negative, allowMissingColumns=True)
    work = work.repartition(SPARK_SQL_SHUFFLE_PARTITIONS).persist(StorageLevel.MEMORY_AND_DISK)
    work.count()

    label_counts = {
        float(r["label"]): int(r["count"])
        for r in work.groupBy("label").count().collect()
    }

    meta = {
        "status": "ok",
        "target_processing_date_hour": str(target_processing_hour),
        "positive_rows": int(label_counts.get(1.0, 0)),
        "negative_rows": int(label_counts.get(0.0, 0)),
        "total_rows": int(sum(label_counts.values())),
        "comparison_hours": [str(h) for h in comparison_hours],
        "negative_info": negative_info,
    }

    return work, meta
```

```python
if RUN_BATCH_MONITORING:
    target_batch_list = pd.date_range(start=BATCH_START, end=BATCH_END, freq=BATCH_FREQ).tolist()

    print("Number of target batches:", len(target_batch_list))
    print("First target batch:", target_batch_list[0])
    print("Last target batch :", target_batch_list[-1])

    batch_auc_results = []

    for i, batch_hour in enumerate(target_batch_list):
        print("\n" + "=" * 80)
        print(f"[{i+1}/{len(target_batch_list)}] Target batch: {batch_hour}")

        batch_df = None
        artifacts = None

        try:
            batch_df, meta = build_dataset_for_target_batch_spark(
                source_df=df_raw,
                target_processing_hour=batch_hour,
                random_state=RANDOM_STATE + i,
            )

            if meta.get("status") != "ok":
                print("Skipped:", meta)
                batch_auc_results.append(meta)
                continue

            need_shap = bool(RUN_BATCH_SHAP)
            result, artifacts = train_synapseml_on_work_df(
                batch_df,
                synapseml_params=selected_synapseml_params,
                random_state=RANDOM_STATE + i,
                shap_col=("featuresShap" if need_shap else None),
                return_artifacts=need_shap,
            )

            row = {**meta, **result}
            batch_auc_results.append(row)
            print("Result:", row)

            # Optional SHAP per batch nếu cần.
            if need_shap and artifacts is not None:
                shap_top_df_spark, _ = compute_top_shap_from_predictions(
                    artifacts["test_pred_df"],
                    feature_cols=artifacts["feature_cols"],
                    shap_col="featuresShap",
                    top_n=10,
                    sample_size=1000,
                    seed=RANDOM_STATE + i,
                )
                shap_top_pdf = shap_top_df_spark.toPandas()
                shap_top_pdf["target_processing_date_hour"] = str(batch_hour)
                display(shap_top_pdf)

        except Exception as e:
            print("Failed:", repr(e))
            batch_auc_results.append({
                "status": "failed",
                "target_processing_date_hour": str(batch_hour),
                "reason": repr(e),
            })

        finally:
            try:
                if batch_df is not None:
                    batch_df.unpersist()
            except Exception:
                pass
            gc.collect()

    auc_series_df = pd.DataFrame(batch_auc_results)
    auc_series_path = f"{OUTPUT_DIR}/batch_auc_series.csv"
    auc_series_df.to_csv(auc_series_path, index=False)

    print("Saved:", auc_series_path)
    display(auc_series_df)
else:
    auc_series_df = pd.DataFrame()
    print("Batch monitoring skipped. Set RUN_BATCH_MONITORING=True to run.")
```

## 23. Empirical p-value for AUC monitoring

Nếu đã có chuỗi AUC theo batch, ta dùng empirical p-value để xem AUC hiện tại có thuộc vùng cao bất thường so với lịch sử trước đó hay không.


```python
def add_empirical_auc_anomaly_score(
    auc_df,
    auc_col="test_auc",
    time_col="target_processing_date_hour",
    min_history_points=20,
    rolling_window=None,
    alpha=0.05,
):
    """
    Tính empirical p-value theo lịch sử trước thời điểm hiện tại.

    Công thức:
    p_value = (số AUC lịch sử >= AUC hiện tại + 1) / (số điểm lịch sử + 1)

    Nếu p_value <= alpha, AUC hiện tại nằm ở vùng cao bất thường so với lịch sử.
    """
    data = auc_df.copy()

    if data.empty or auc_col not in data.columns:
        return data

    data = data.sort_values(time_col).reset_index(drop=True)

    p_values = []
    percentile_ranks = []
    history_means = []
    history_medians = []
    history_qs = []
    history_sizes = []
    anomaly_flags = []
    anomaly_reasons = []

    for i, row in data.iterrows():
        current_auc = row.get(auc_col, np.nan)
        history = data.loc[:i-1, auc_col].dropna()

        if rolling_window is not None:
            history = history.tail(int(rolling_window))

        history_size = len(history)
        history_sizes.append(history_size)

        if history_size < min_history_points or pd.isna(current_auc):
            p_values.append(np.nan)
            percentile_ranks.append(np.nan)
            history_means.append(np.nan)
            history_medians.append(np.nan)
            history_qs.append(np.nan)
            anomaly_flags.append(False)
            anomaly_reasons.append("warmup_not_enough_history")
            continue

        p_value = (np.sum(history >= current_auc) + 1) / (history_size + 1)
        percentile_rank = (np.sum(history <= current_auc) + 1) / (history_size + 1)

        history_mean = history.mean()
        history_median = history.median()
        history_q = history.quantile(1 - alpha)

        is_anomaly = p_value <= alpha

        p_values.append(float(p_value))
        percentile_ranks.append(float(percentile_rank))
        history_means.append(float(history_mean))
        history_medians.append(float(history_median))
        history_qs.append(float(history_q))
        anomaly_flags.append(bool(is_anomaly))

        if is_anomaly:
            anomaly_reasons.append(f"auc_is_in_top_{int(alpha * 100)}pct_tail_of_history")
        else:
            anomaly_reasons.append("normal_vs_history")

    q_col = f"history_auc_q{int((1-alpha)*100)}"
    data["history_size"] = history_sizes
    data["history_auc_mean"] = history_means
    data["history_auc_median"] = history_medians
    data[q_col] = history_qs
    data["auc_empirical_p_value"] = p_values
    data["auc_percentile_rank"] = percentile_ranks
    data["is_auc_anomaly"] = anomaly_flags
    data["auc_anomaly_reason"] = anomaly_reasons

    return data


if len(auc_series_df) > 0:
    auc_monitoring_df = add_empirical_auc_anomaly_score(
        auc_series_df,
        auc_col="test_auc",
        time_col="target_processing_date_hour",
        min_history_points=MIN_HISTORY_POINTS,
        rolling_window=ROLLING_WINDOW,
        alpha=ALPHA,
    )

    auc_monitoring_path = f"{OUTPUT_DIR}/batch_auc_empirical_pvalue_monitoring.csv"
    auc_monitoring_df.to_csv(auc_monitoring_path, index=False)

    print("Saved:", auc_monitoring_path)
    display(auc_monitoring_df)
else:
    print("No AUC series available. Run batch monitoring first.")
```

```python
if "auc_monitoring_df" in globals() and len(auc_monitoring_df) > 0:
    plot_df = auc_monitoring_df.copy()
    plot_df["target_processing_date_hour"] = pd.to_datetime(plot_df["target_processing_date_hour"])
    plot_df = plot_df.sort_values("target_processing_date_hour")

    plt.figure(figsize=(12, 5))
    plt.plot(plot_df["target_processing_date_hour"], plot_df["test_auc"], marker="o", label="test_auc")

    q_col = f"history_auc_q{int((1-ALPHA)*100)}"
    if q_col in plot_df.columns:
        plt.plot(plot_df["target_processing_date_hour"], plot_df[q_col], linestyle="--", label=q_col)

    if "is_auc_anomaly" in plot_df.columns:
        anomalies = plot_df[plot_df["is_auc_anomaly"] == True]
        if len(anomalies) > 0:
            plt.scatter(anomalies["target_processing_date_hour"], anomalies["test_auc"], s=80, label="anomaly")

    plt.xlabel("target_processing_date_hour")
    plt.ylabel("AUC")
    plt.title("Batch AUC Monitoring")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No AUC monitoring plot.")
```
